## Variable Metadata

```python
variable_metadata = {}
for dataset_id in metadata['dataset_id']:
    result = get_.variable_metadata(datamart_api_url, dataset_id)
    if result is None:
        print('Error: getting variable metadata:', dataset_id)
    else:
        variable_metadata[dataset_id] = result
```

```python
var_count = []
for dataset_id, vmetadata in variable_metadata.items():
    var_count.append({'dataset_id': dataset_id, 'num_variables': vmetadata.shape[0]})
display(pd.DataFrame(var_count))
```

```python
for dataset_id, vmetadata in variable_metadata.items():
    print(dataset_id)
    display(vmetadata)
```

## Variable Canonical CSVs

```python
missing_variables = []
error_getting_data = []
zero_length_data = []
for dataset_id, vmetadata in variable_metadata.items():
    if vmetadata.shape[0] == 0:
        missing_variables.append(dataset_id)
        continue
    variable_id = vmetadata.loc[0, 'variable_id']
    data = get_.variable_data(datamart_api_url, dataset_id, variable_id)
    if data is None:
        error_getting_data.append((dataset_id, variable_id))
        continue
    if data.shape[0] == 0:
        zero_length_data.append((dataset_id, variable_id))
        continue
    print(dataset_id, variable_id)
    display(data)
```

```python
print('Datasets with missing variables:', missing_variables)
```
